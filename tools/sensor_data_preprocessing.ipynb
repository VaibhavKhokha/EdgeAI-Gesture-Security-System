```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
Idle = pd.read_csv('../Idle.csv')
Gesture = pd.read_csv('../Gesture.csv')
```

```python
print(Idle.head())
print(Idle.shape[0])
print(Idle.shape[1])
```

```python
print(Gesture.head())
print(Gesture.shape[0])
print(Gesture.shape[1])
```

```python
#dropping last row to match both datasets
Idle = Idle.drop(Idle.index[-1])
print(Idle.head())
print(Idle.shape[0])
print(Idle.shape[1])
```

```python
physical_gestures = 32
snapshots = 40
readings = 6
Idle = np.array(Idle)
Gesture = np.array(Gesture)

#converting to 3d matrix
Idle = Idle.reshape(physical_gestures, snapshots, readings)
Gesture = Gesture.reshape(physical_gestures, snapshots, readings)

print(Idle)
print(Idle.shape[0])
print(Idle.shape[1])

print(Gesture)
print(Gesture.shape[0])
print(Gesture.shape[1])
```

```python
#made targets for both sets
target_idle = np.zeros(32)
target_gesture = np.ones(32)
```

```python
#joined both targets and sets along the x-axis just stacked on to of each other
X = np.concatenate((Idle, Gesture), axis = 0)
Y = np.concatenate((target_idle, target_gesture), axis = 0)
```

```python
#shuffling these both in the same fashion
shuffled_indices = np.random.permutation(len(X))

X = X[shuffled_indices]
Y = Y[shuffled_indices]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size = 0.2, random_state = 42, stratify = Y)
```

```python
np.savez(
    'kinematic_dataset.npz', X_train = X_train, X_test = X_test, Y_train = Y_train, Y_test = Y_test)
```
