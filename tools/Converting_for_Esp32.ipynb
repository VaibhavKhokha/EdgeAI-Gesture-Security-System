```python
import numpy as np
import tensorflow as tf
import os
import re
import sys
```

```python
def file_to_cpp(input_path, output_path = None):
    #checking if file exists
    if not os.path.exists(input_path):
        print(f"Error: File '{input_path}' not found..", file = sys.stderr)
        return

    #generating a safe variable name
    base_name = os.path.basename(input_path)
    var_name = re.sub(r'[^a-zA-Z0-9]', '_', base_name)

    #reading binary content
    with open(input_path, 'rb') as file:
        data = file.read()

    data_len = len(data)

    lines = [f"unsigned char {var_name}[] = {{"]

    #grping into blocks of 12 hex entries per row to match xxd output
    for i in range(0, data_len, 12):
        chunk = data[i : i+12]
        hex_row = ", ".join(f"0x{byte:02x}" for byte in chunk)

        #adding comma at end if it is not final row block
        if i + 12 < data_len:
            lines.append(f"  {hex_row},")
        else:
            lines.append(f"  {hex_row}")

    lines.append("};")
    lines.append(f"unsigned int {var_name}_len = {data_len};")

    header_content = "\n".join(lines) + "\n"

    #saving file
    if output_path:
        with open(output_path, 'w', encoding = 'utf-8') as file:
            file.write(header_content)
            print(f"Successfull")
    else:
        print(header_content)
```
